```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
df = pd.read_csv ('tips.csv')
df_binary = df[['total_bill', 'tip']]
df_binary.columns = ('Bill', 'Tip')
x = np.array(df_binary['Bill']).reshape(-1, 1)
y = np.array(df_binary['Tip']).reshape(-1, 1)
regr = LinearRegression()
regr.fit (x, y)
print ('slope of the line is',regr.coef_)
y_pred = regr.predict (x)
plt.scatter(x, y, color ='b')
plt.plot(x, y_pred, color ='k')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt. title('Linear Regression')
```

```python
x_trial=19.89
xfinal= np.array(x_trial).reshape (-1, 1)
y_trial = regr.predict (xfinal)
print ('predicted tip for $',x_trial,'is',y_trial[0][0])
plt.show()
```
