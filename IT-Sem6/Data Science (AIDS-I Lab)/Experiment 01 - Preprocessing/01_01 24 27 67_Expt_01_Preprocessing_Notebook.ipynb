## Setting index column

```python
df = source
df.reset_index(level=0, inplace=True)
df.head()
```

## Handling Missing and Duplicate Data

```python
df.isnull()
```

```python
# Finding out number of duplicated values
df.duplicated().sum()
```

```python
# Finding out missing values
# ld.isnull().sum()
# or we can use isna()
df.isna().sum()
```

```python
#Filling the missing data in numerical columns with its mean
df = df.fillna(df.mean())
print(df.shape)
print(df.isna().sum())
```

```python
df["Market Category"].unique().size
```

```python
#Since the attribute Market Category does not affect the MSRP, we will drop this column
df.drop(columns = ["Market Category"], inplace = True)
print(df.shape)
print(df.isna().sum())
```

```python
#Dropping the missing data present due to non-numerical and non-categorical columns
df = df.dropna()
print(df.shape)
print(df.isna().sum())
```

## Outliers

```python
#summary statistics of all the columns
df.describe()
```

```python
# If we compare the mean of each column with the min/max value, we'll notice that engine_hp, highway_mpg, city_mpg and popularity might have outliers as there's a considerable difference between average value and max value
```

```python
#Checking for outliers
plt.figure(figsize=(16,8)) #(width,height)
sns.boxplot(x='Engine HP', data=df);
```

```python
plt.figure(figsize=(16,8))
sns.histplot(df['Engine HP'],kde=True, stat="density", linewidth=0)

#from this we got to know that we have to use IQR based filtering because our data distribution is skewed
```

```python
#Finding Quartiles and IQR
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
IQR["Engine HP"], Q1["Engine HP"], Q3["Engine HP"]
```

```python
# calculate the outlier cutoff based on Engine HP
cut_off = IQR["Engine HP"] * 1.5
lower, upper = Q1["Engine HP"] - cut_off, Q3["Engine HP"] + cut_off
```

```python
# getting the outliers
outliers = pd.DataFrame([v for k,v in df.iterrows() if v["Engine HP"] < lower or v["Engine HP"] > upper])
outliers.head(10)
```

```python
outliers.shape
```

```python
# dataframe after removing outliers based on Engine HP
df = pd.DataFrame([v for k,v in df.iterrows() if v["Engine HP"] >= lower and v["Engine HP"] <= upper])
```

```python
df.shape
```

```python
plt.figure(figsize=(16,8))
sns.histplot(df['Engine HP'],kde=True, stat="density", linewidth=0)
```

```python
# Finding 10 largest records with the largest value in Engine HP
df.nlargest(10,["Engine HP"])
```

```python
from sklearn.preprocessing import LabelEncoder

encoderOne = LabelEncoder()

#here encoderOne transforms both the attributes
df["Vehicle Style"] = encoderOne.fit_transform(df["Vehicle Style"])
df["Model"] = encoderOne.fit_transform(df["Model"])
```

```python
from sklearn.preprocessing import OneHotEncoder

encoderTwo = OneHotEncoder(sparse=False)
 
encodedTT = pd.DataFrame(encoderTwo.fit_transform(df[["Transmission Type", "Vehicle Size"]]))
encodedTT.columns = encoderTwo.get_feature_names(["Transmission Type", "Vehicle Size"]) 
encodedTT.head()
```

```python
df = pd.get_dummies(df, columns=["Driven_Wheels", "Transmission Type", "Vehicle Size"], drop_first=True)
df.head()
```
